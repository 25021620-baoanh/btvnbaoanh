```python
#w6a2
def tong(arr):
    result = [arr[0]]
    for i in range(1, len(arr)):
        result.append(result[-1] + arr[i])
    return result
print(tong(list(map(int,input().split()))))
```

```python
#w6a4
def a(arr):
    rec=dict()
    for i in range(len(arr)):
        arr[i] = arr[i].split(":")
        if arr[i][0] not in rec:
            rec[arr[i][0]] =[arr[i][1]]
        else:
            rec[arr[i][0]].append(arr[i][1])
    return rec
a(input().split())
```

```python
#w6a10
def sum(arr,k):
    re=[]
    for i in range(len(arr)):
        for j in arr[i +1:]:
            if arr[i] + j ==k:
                re.append((arr[i],j))
    return sorted(re)
sum(list(map(int,input().split())) , int(input()))
 
```
