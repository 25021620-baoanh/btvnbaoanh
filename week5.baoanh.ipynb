```python
#w5a1
def max(a,b):
    if a<b:
        return b
    else:
        return a
a,b=map(int, input().split())
print(max(a,b))
```

```python
#w5a2
def doi(a,b):
    return b,a
a,b=map(int,input().split())
print(doi(a,b))
```

```python
#w5a3
def check(a):
    if a==2:
        return True
    for i in range (2,a):
        if a%i==0:
            return False
    return a>1
a=int(input())
print(check(a))
```

```python
#w5a4
def shh(a):
    s=0
    for i in range(1,a):
        if a%i==0:
            s+=i
    if s==a:
        return True
    else:
        return False
a=int(input())
print(shh(a))
```

```python
#w5a7
a=input().split()
if a[1]=="+":
    print(round(float(a[0])+float(a[2]),2))
elif a[1]=="-":
    print(round(float(a[0])-float(a[2]),2))
elif a[1]=="*":
    print(round(float(a[0])*float(a[2]),2))
elif a[1]=="/" and int(a[2])!=0:
    print(round(float(a[0])/float(a[2]),2))
else:
    print(False)



```

```python
#w5a8
def ham(a,b):
    k=bin(a^b)
    s=0
    for i in k:
        if i=="1":
            s=s+1
    return s
a,b=map(int,input().split())
print(ham(a,b))
```

```python
#w5a9
def tong(a):
    s=0
    while a!=0:
        s=s+a%10
        a=a//10
    return s
a=int(input())
print(tong(a))



```
