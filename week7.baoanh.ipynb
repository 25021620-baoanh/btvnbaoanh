```python
#w7a1
def binarySearch(arr,k):
    left=0
    right=len(arr)
    while right >left:
        mid=(right+left)//2
        if arr[mid] > k:
            right = mid-1
        elif arr[mid]< k:
            left =mid +1
        else:
            return mid
    return -1
binarySearch(list(map(int,input().split())),int(input()))
```

```python
#w7a3
def BubbleSort(arr):
    for i in range(len(arr)):
        check= False
        for j in range(0,len(arr)-1-i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]= arr[j+1],arr[j]
                check= True
        if not check:
            break
    return arr
BubbleSort(list(map(int, input().split())))
def selection_sort(a):
    selec=[]
    for i in range(len(a)):
        selec.append(min(a))
        a.remove(min(a))
    return selec
selection_sort(list(map(int,input().split())))
            
```

```python
#w7a5
def count_pairs(a, X):
    count = 0
    seen = {}
    for i in a:
        target = X - i
        if target in seen:
            count += seen[target]
        seen[i] = seen.get(i, 0) + 1
    return count
count_pairs(list(map(int,input().split())),int(input()))
```

```python
#w7a7
a=list(map(int, input().strip()[1:-1].split()))
n=len(a)
tail=[min(a)-1] + [max(a)+1]*n
ans =0
for i in range(n):
    l,r=0,i
    while l <=r:
        mid =(l+r)//2
        if tail[mid] < a[i]:
            l = mid +1
        else:
            r= mid -1
    if r>=0 and tail[r] < a[i]:
        ans =max(ans, r+1)
        tail[r+1] = min(tail[r+1], a[i])
print(ans) 





    
```
