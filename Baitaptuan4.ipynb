```python
num = int(input("nhập số:"))

flag = False
count = 0

while flag == False:
    str_num = str(num)
    len_num = len(str_num)
    id = len_num - 1
    reverse_num = ""

    for i in range(len(str_num) - 1, -1, -1):
        reverse_num = reverse_num + str_num[i]

    reverse_num = int(reverse_num)

    for i in range(len_num // 2):
        if str_num[i] != str_num[id]:
            count += 1
            flag = False
            break
        flag = True
        id -= 1
    
    if flag == False:
        num += reverse_num

print(f"{count} {num}")
#b8
```
